```python
from pymongo import MongoClient
import pandas as pd
client = MongoClient()
db = client.openfoodfacts
collection = db.products

import sys
sys.path.append('../')
from func import *
```

```python
EAN13_temp = pd.read_csv('../M_MERGE/output_data/M_PROD_FINAL.csv')['O_EAN13']
EAN13_temp = EAN13_temp.astype(str).values.flatten()

O_EAN13 = []
for ean in EAN13_temp :
    O_EAN13.append(ean)
```

```python
len(set(O_EAN13))
```

```python
proj = {
    'product_name_fr' : 1, 
    'ingredients' : 1,
    '_id' : 1,
}
```

```python
match = { '_id' : { '$in' : list(O_EAN13) } }
```

```python
c_f = collection.aggregate([
    { '$match' : match } ,
    { '$project' : proj },
])

ingredients = pd.DataFrame(list(c_f))
ingredients.rename(columns={'_id' : 'O_EAN13'}, inplace=True)
ingredients['ingredients'][ingredients['ingredients'].isna()] = [{}]
```

```python
ingredients['ingredients'][0]
```

```python
len(ingredients['O_EAN13'].unique())
```

```python
id_ = []
denomination = []
ing = []
has_sub_ingredients = []
sub_ingredients = []
text = []
percent_min = []
percent_max = []
percent = []
rank = []
product_name_fr = []
O_EAN13 = []

for i in range(len(ingredients['ingredients'])) :
    for dict_ in ingredients['ingredients'][i] : 
        
        product_name_fr.append(ingredients['product_name_fr'][i])
        O_EAN13.append(ingredients['O_EAN13'][i])
        
        if 'id' in dict_.keys() :
            id_.append(dict_['id'])
        else :
            id_.append(None)
        if 'ingredients' in dict_.keys() :
            ing.append(dict_['ingredients'])
        else :
            ing.append(None)
        if 'has_sub_ingredients' in dict_.keys() :
            has_sub_ingredients.append(dict_['has_sub_ingredients'])
        else :
            has_sub_ingredients.append(None)
        if 'ingredients' in dict_.keys() :
            sub_ingredients.append(dict_['ingredients'])
        else :
            sub_ingredients.append(None)
        if 'text' in dict_.keys() :
            text.append(dict_['text'])
        else :
            text.append(None)
        if 'percent_min' in dict_.keys() :
            percent_min.append(dict_['percent_min'])
        else :
            percent_min.append(None)
        if 'percent_max' in dict_.keys() :
            percent_max.append(dict_['percent_max'])
        else :
            percent_max.append(None)
        if 'percent' in dict_.keys() :
            percent.append(dict_['percent'])
        else :
            percent.append(None)
        if 'rank' in dict_.keys() :
            rank.append(dict_['rank'])
        else :
            rank.append(None)
    
```

```python
df_ingredients = pd.DataFrame(
    list(zip(O_EAN13, id_,  has_sub_ingredients, sub_ingredients, text, percent_min, percent_max, percent, rank, product_name_fr)), 
    columns =['O_EAN13', 'id_ingredient', 'has_sub_ingredients', 'sub_ingredients', 'text', 'percent_min', 'percent_max', 'percent', 'rank', 'product_name_fr'])
```

```python
def label_ingredient (row):
    if row['denomination'][:2] == 'en' :
        return 0
    return 1
```

```python
df_ingredients['unknown_ingredient'] = df_ingredients.apply(lambda row: label_ingredient(row), axis=1)
```

```python
df_ingredients.to_csv('O_ingredients.csv')
```
