```python
# CHECK THAT ALL EAN13 CONTAIN ONLY 13 CHARACTERS
pb_length=[]
for i in N_EAN13:
    if len(i)!=13:
        pb_length.append(i)
        print(' length:',len(i),':',i)
        #pb_index.append(N_EAN13.index(i))
        #print('index:',N_EAN13.index(i),' length:',len(i),':',i)
print('\n',len(pb_length),' EAN13 codes are not 13 character long')
```

```python
# CHECK THAT ALL EAN13 CONTAIN ONLY NUMERICAL VALUES
pb=[]
pb_index=[]
for i in N_EAN13:
    if i.isdecimal()==False:
        pb.append(i)
        print(' length:',len(i),':',i)
        #pb_index.append(N_EAN13.index(i))
        #print('index:',N_EAN13.index(i),' length:',len(i),':',i)
print('\n',len(pb),' EAN13 codes are not numeric values (problem)')
if pb==pb_length:
    print('\nall issues with length and nunmerical values relate to the same ITEMS')
```

```python
for p in pb:
    print('\nEAN13 :',p)
    print(data[data['ITEM']==p][['STD/PROMO','MARQUE','NIELSEN_DESCRIPTION']])
    
    
    
```

```python
# TOUS LES EAN13 A 16 CHIFFRES CORRESPONDENT A DES PROMOS 
# 136 occurences = 8 markets x 17 KPIs ('FACT')
# transformation des EAN13 TROP LONG EN 13 CHIFFRES (suppression des 3 chiffres en trop)
# Creation of a new column with only true EAN13

for i in range(0,data.shape[0]):
    if data.iloc[i,10] in pb:
       data.iloc[i,-1]=data.iloc[i,10][0:13] #selection of 13 first caracters
    if data.iloc[i,10] in pb_length:
       data.iloc[i,-1]=data.iloc[i,10][0:13] #selection of 13 first caracters    
```

```python
#création nouvelle liste des EAN13 (ancienne N_EAN13)
EAN13=data['EAN13'].unique()
print(EAN13[:20])
print(EAN13[-20:])

EAN13=EAN13[1:]

#vérification que tous les EAN113 ont bien 13 charactères et ne contiennent que des chiffres
pb_length=[]
for i in EAN13:
    if len(i)!=13:
        pb_length.append(i)
        print(' length:',len(i),':',i)
        #pb_index.append(N_EAN13.index(i))
        #print('index:',N_EAN13.index(i),' length:',len(i),':',i)
print('\n',len(pb_length),' EAN13 codes are not 13 character long')

#vérification que tous les EAN13 ne contienne que des chiffres
pb=[]
for i in EAN13:
    if i.isdecimal()==False:
        pb.append(i)
        print(' length:',len(i),':',i)
print('\n',len(pb),' EAN13 codes are not numeric values (problem)')

```
