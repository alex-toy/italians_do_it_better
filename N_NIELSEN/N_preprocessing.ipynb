```python
#-------------------------------------------------------------------------------
#-------------------------------------------------------------------------------
#  4.2. EXTRACT ORGANIC PRODUCTS (BIO) FROM NIELSEN DESCRIPTION   #
#  (BIO / NON BIO)                                                #
```

```python
# ABC LAMBDA FUNCTION TO SPLIT ON '*'
#data['ORGANIC'] = data['NIELSEN_DESCRIPTION']+'*'
#data['ORGANIC'] = data['NIELSEN_DESCRIPTION'].apply(lambda desc: desc.split('*')[1:])
#data['ORGANIC'] = data['NIELSEN_DESCRIPTION'].str.replace(data['ORGANIC'],'')
#print(data['ORGANIC'])
```

```python
def split_star_left(col) :
    if isinstance(col, str) :
        return col.split('*', 1)[0] # keep first part [0] of split (before the splitter (st[1])
    else :
        return ''

def split_star_right(col) :
    if isinstance(col, str) :
        split = col.split('*', 2)
        if len(split)>2:
            return split[2]  # keep second part [1] of split (after the splitter (st[1])
        elif len(split)>1:
            return split[1]  # keep second part [1] of split (after the splitter (st[1])
        else:
            return 'STAR not found'
    else :
        return ''
```

```python
def bio_non_bio(col) :
    if isinstance(col, str) :
        if col[-7:]=='NON BIO':
            return ' NON BIO'
        elif (col[-3:]=='BIO')&(col[-7:]!='NON BIO'):
            return ' BIO'
        else:
            return 'separator'
    else :
        return ''
    print(n)
```

```python
# ORGANIC DEFINITION BASED ON 'BIO' and 'NON BIO'
data['ORGANIC'] = list(map(bio_non_bio, data['NIELSEN_DESCRIPTION']))

print(set(data['ORGANIC']))
print(data[['ORGANIC','EAN13','MARQUE']].groupby(by='ORGANIC').count())
temp=data[data['ORGANIC']=='unknown']
print('\nnumber of EAN13 products not recognised:',len(temp['EAN13'].unique()))
```

```python
#-------------------------------------------------------------------------------
#-------------------------------------------------------------------------------
#     4.3. PROCESSING WEIGHT                                        #
#     (STRING / NON STRING)                                       #
```

```python
#-------------------------------------------------------------------------------
# EXTRACT WEIGHT (STRING) FROM NIELSEN_DESCRIPTION

import re

# s is the search to execute (compile)
 

def find_weight(text):
    s = re.compile('\d+G')
    if isinstance(text,str):
        weights=s.findall(text)
        if len(weights) >0:
            return weights[-1]
    
data['WEIGHT']=data['NIELSEN_DESCRIPTION'].apply(lambda row: find_weight(row))
```

```python
WEIGHT_U=data['WEIGHT'].unique()
print('number of different weights in database:',len(WEIGHT_U))
WEIGHT_U
```

```python
#-------------------------------------------------------------------------------
# EXTRACT WEIGHT AS INTEGER FROM WIEGHT (string)

def remove_G_in_weight(text):
    if isinstance(text,str):
        return text[:-1]
    else:
        return '0'
   
data['WEIGHT_num']=data['WEIGHT'].apply(lambda row: remove_G_in_weight(row)).astype(int)
print(sorted(data['WEIGHT_num'].unique()))

#!!!!!!!!!!!!!!  0 = UNKOWN !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
```
