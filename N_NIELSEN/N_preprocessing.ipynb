```python
#FUNCTION TO CHECK UNIQUE NON-NULL AND NON'' VALUES IN A COLUMN
def unique_clean(series):
    print('Unique (non null) values in',series.name,':\n')
    u=series.unique()
    clean=[]
    for m in u:
        if (pd.isna(m)==False and m!=''):
            clean.append(m)
            print(m)
    print('\nnumber of ',series.name,': ',len(clean))
    print(type(clean),'\n')
    
    return clean
```
