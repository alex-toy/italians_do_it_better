```python
import pandas as pd
import numpy as np
import re
```

```python
df_N_FR_FEC = pd.read_csv('N_NIELSEN/N_FRANCE_FEC.csv')
```

```python
#Création colonnes N_DESC_SHORT et LONG_BRAND
#Temporaire
df_N_FR_FEC['N_DESC_SHORT'] = df_N_FR_FEC['NIELSEN_DESCRIPTION'].copy()
df_N_FR_FEC['N_DESC_SHORT'] = df_N_FR_FEC['N_DESC_SHORT'].str.replace('SAUCES POUR FECULENTS ', '', regex=True, case=False)
df_N_FR_FEC['N_DESC_SHORT'] = df_N_FR_FEC['N_DESC_SHORT'].str.replace('POUR FECULENTS ', '', regex=True, case=False)
df_N_FR_FEC['N_DESC_SHORT'] = df_N_FR_FEC['N_DESC_SHORT'].str.replace('FECULENTS ', '', regex=True, case=False)

LONG_BRAND = []
for k in range(len(df_N_FR_FEC['N_DESC_SHORT'])) :
    temp = df_N_FR_FEC['N_DESC_SHORT'][k].replace(df_N_FR_FEC['EMBALLAGE'][k], "").strip()
    df_N_FR_FEC['N_DESC_SHORT'][k] = temp
    temp = df_N_FR_FEC['N_DESC_SHORT'][k].split(df_N_FR_FEC['GAMME'][k], 1)
    LONG_BRAND.append(temp[0].strip())
    

df_N_FR_FEC['LONG_BRAND'] = LONG_BRAND
```
